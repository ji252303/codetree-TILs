(x, y) 위치에서 시작하여 한 칸 이동하려 합니다.<br>
숫자 0이 주어지면 동쪽으로,<br> 
숫자 1이 주어지면 남쪽으로,<br>
숫자 2가 주어지면 서쪽으로, <br>
숫자 3이 주어지면 북쪽으로 이동하려 합니다.

```python
dir_num = 2 # 주어진 방향이 서쪽인 경우
x, y = 1, 5 # 현재 위치가 (1, 5)인 경우

if dir_num == 0:
    nx, ny = x + 1, y
elif dir_num == 1:
    nx, ny = x, y - 1
elif dir_num == 2:
    nx, ny = x - 1, y
else:
    nx, ny = x, y + 1
```

이게 가능한 이유는, 각 방향에 맞는 x좌표의 차를 dx에 y좌표의 차를 dy에 적어주었기 때문입니다.<br>
예를 들어 dir_num이 2인 경우에는 x좌표만 1이 감소하기 때문에 dx[2] = -1, dy[2] = 0이 됩니다.

```python
dir_num = 2 
x, y = 1, 5
dx, dy = [1, 0, -1, 0], [0, -1, 0, 1]

if dir_num == 0:
    nx, ny = x + dx[0], y + dy[0]
elif dir_num == 1:
    nx, ny = x + dx[1], y + dy[1]
elif dir_num == 2:
    nx, ny = x + dx[2], y + dy[2]
else:
    nx, ny = x + dx[3], y + dy[3]
```

```python
dir_num = 2 
x, y = 1, 5
dx, dy = [1, 0, -1, 0], [0, -1, 0, 1]

nx, ny = x + dx[dir_num], y + dy[dir_num]
```
